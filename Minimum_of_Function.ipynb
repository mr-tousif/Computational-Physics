- Write a function that determines the height and radius of the cylinder that has the minimum surface area for a given volume.



Volume $\mathbf{V=\pi r^2 h}$  and  Surface area is $\mathbf{S=2\pi r h+2 \pi r^2}$ for finding the minimum surface area of the cylinder we need to solve second derivative of the S.

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.optimize 

V=100 #for any given volume
def Area(r):
    Area = 2*V/r+2*np.pi*r**2 
    return Area

def Volume(radius):
    h=V/(np.pi*radius**2)

r=np.linspace(1,10,100)
plt.plot(r,Area(r),"mo")
plt.xlabel("r")
plt.ylabel("Area")
print("The optimal Solution is:")
print("Area:", min(Area(r)))
```

```python
radius=r[np.argmin(Area(r))] #radius
print("radius of the cylinder is", radius)
```

```python
Volume(radius)
print("Height of Cylinder for Minimum Surface Area is", h)
```

```python
result_2=optimize.minimize_scalar(Area, bounds=(1,10), method ="Bounded")
```

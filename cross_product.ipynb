- Define a function that computes the cross product of two vectors: $\mathbf{C}=\mathbf{A}\times\mathbf{B}$. Make sure that the function works even if one or both of $\mathbf{A}$ and $\mathbf{B}$ are arrays of vectors. Document the function properly.
- Plot the three components of $\mathbf{C}(x)=\mathbf{A}(x)\times\mathbf{B}$ as functions of $x$ in $[0,10]$, where $\mathbf{A}(x)= [\cos(x), \sin(x), 0]$ is a vector field, and $\mathbf{B}=[1, 0, 1]$ is a constant vector. Label the axes and curves properly.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def A(x):
    return np.array([np.cos(x),np.sin(x),0])
    
B = np.array([1,0,1])

def C(x): 
    return np.cross(A(x),B)
    
```

```python
x = np.arange(11)
x
```

```python
Cross = []
for i in x:
    Cross.append(C(x[i]))
```

```python
Cross
```

```python
Cross_i_component = []

for i in x:
    Cross_i_component.append(Cross[i][0])
```

```python
Cross_j_component = []

for i in x:
    Cross_j_component.append(Cross[i][1])
```

```python
Cross_k_component = []

for i in x:
    Cross_k_component.append(Cross[i][2])
```

```python
plt.plot(x,Cross_i_component,label="x component")
plt.plot(x,Cross_j_component,label="y component")
plt.plot(x,Cross_k_component,label="z component")
plt.legend()
plt.xlabel("x")
plt.ylabel(f"A(x)\cross(B)")
plt.show()
```
